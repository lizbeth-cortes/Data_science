```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta_1 = 'https://github.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/raw/master/Datasets/near_earth_objects-jan_feb_1995-clean.csv'
ruta_2 = 'https://github.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/raw/master/Datasets/athlete_olympic_events-clean-no_repetition.csv'
ruta_3 = 'https://github.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/raw/master/Datasets/videogame_sales-raw.csv'
```

```python
df_1 = pd.read_csv(ruta_1, index_col=0)

df_1.head()
```

```python
sns.scatterplot(x=df_1['estimated_diameter.meters.estimated_diameter_min'], y=df_1['estimated_diameter.meters.estimated_diameter_max']);
df_1['estimated_diameter.meters.estimated_diameter_min'].corr(df_1['estimated_diameter.meters.estimated_diameter_max'])
```

```python
sns.scatterplot(x=df_1['estimated_diameter.meters.estimated_diameter_max'], y=df_1['relative_velocity.kilometers_per_second']);
df_1['estimated_diameter.meters.estimated_diameter_max'].corr(df_1['relative_velocity.kilometers_per_second'])
```

```python
df_2 = pd.read_csv(ruta_2, index_col=0)

df_2.head()
```

```python
sns.jointplot(x=df_2['age'], y=df_2['weight'], kind='hex');
df_2['age'].corr(df_2['weight'])
```

```python
sns.jointplot(x=df_2['height'], y=df_2['weight'], kind='hex');
df_2['height'].corr(df_2['weight'])
```

```python
df_4 = pd.read_csv(ruta_3)

df_4.head()
```

```python
df_4 = df_4[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
```

```python
df_4 = df_4[df_4 > 0]
df_4 = df_4[df_4['NA_Sales'] < 20]
df_4 = df_4[df_4['JP_Sales'] < 10]
```

```python
sns.scatterplot(x=df_4['NA_Sales'], y=df_4['JP_Sales']);
df_4['NA_Sales'].corr(df_4['JP_Sales'])
```

```python
sns.scatterplot(x=df_4['Other_Sales'], y=df_4['Global_Sales']);
df_4['Other_Sales'].corr(df_4['Global_Sales'])
```

```python
sns.scatterplot(x=df_4['NA_Sales'], y=df_4['Global_Sales']);
df_4['NA_Sales'].corr(df_4['Global_Sales'])
```

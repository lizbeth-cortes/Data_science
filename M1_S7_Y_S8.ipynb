## Ejemplo 4: apply

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data science/Datasets/new_york_times_bestsellers-dirty.csv', index_col=0)

df.head()
```

```python
# La función apply en Pandas es una herramienta poderosa y flexible que permite aplicar
# una función a lo largo de un eje de un DataFrame o a elementos individuales de una Serie.
```

```python
def years_since_bestseller(value):

    as_datetime = pd.to_datetime(value, unit='ms')
    today = pd.to_datetime('today')
    difference_in_days = (today - as_datetime).days
    in_years = difference_in_days / 365

    return in_years
```

```python
df['published_date.numberLong'].apply(years_since_bestseller)
```

```python
# Definimos un DataFrame simple
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
```

```python
# Definimos una función que podemos aplicar usando 'apply'
def add_one(x):
    return x + 1
```

```python
# Usamos 'apply' para aplicar 'add_one' a la columna 'A'
df['A_plus_one'] = df['A'].apply(add_one)
```

```python
df.head()
```

```python
# Usamos una función lambda para realizar una operación similar en la columna 'B'
df['B_plus_one'] = df['B'].apply(lambda x: x + 1)
```

```python
df.head()
```

```python
# Asignamos una funcion lambda a una variable
x = lambda test : test + 10
print(x(5))
```

```python
print(x(167593))
```

## Ejemplo 3: Uniendo DataFrames con merge

```python
import pandas as pd
```

```python
users = pd.read_csv('/content/drive/MyDrive/Data science/Datasets/MovieLens/users-raw.csv',
                    index_col=0, names=['gender', 'age', 'occupation', 'cp'], skiprows=1)
```

```python
users.head()
```

```python
users.index.name = 'user_id'
```

```python
users.head()
```

```python
occupations = pd.read_csv('/content/drive/MyDrive/Data science/Datasets/MovieLens/occupations-raw.csv', index_col=0, names=['description'], skiprows=1)
occupations.index.name = 'occupation_id'
occupations.head()
```

```python
users_full = pd.merge(users, occupations, left_on='occupation', right_index=True).sort_index()
```

```python
users_full.head()
```

```python
users_full = users_full.rename(columns={'occupation': 'occupation_id', 'description': 'occupation'})
users_full.head()
```
